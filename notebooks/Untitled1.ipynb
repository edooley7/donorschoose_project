```python
from __future__ import division
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
import io
```

```python
df = pd.DataFrame(pickle.load(open('clean_2015_data.pkl', 'rb')))
```

```python
school_df = df[0:100]
school_df
```

```python
def get_school_counts(row, df):
    filter = (df["_schoolid"] == row["_schoolid"])  & (df["date_completed"] < row["date_completed"])
    dc_count = df[filter]["date_completed"].count()
    print row['_schoolid']
    return pd.Series([dc_count])
```

```python
%%time 
school_df[["school_previous_completed"]] = school_df.apply(lambda x : get_school_counts(x, school_df),axis=1)
```

```python
school_df
```

```python
%%time 
gg = school_df.groupby("_schoolid")  
df_ = pd.DataFrame()
for g, df in gg:
    df['school_previous_success'] = df.date_posted.map(lambda x: len(df[df.date_completed < x]))
    df['school_previous_projects'] = df.date_posted.map(lambda x: len(df[df.date_posted < x]))
    df['school_per_success'] = df.school_previous_success/df.school_previous_projects
    df_ = pd.concat([df_, df])
```

```python
%%time
posted = school_df.set_index('date_posted').groupby('_schoolid').cumcount()
print posted

school_df['school_previous_posted'] = posted.values
```
